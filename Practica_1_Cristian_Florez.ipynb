l) 	Genere una tercera señal que sea la suma de estas dos señales, es decir s[n]=y[n]+y2[n] y una cuarta, que sea el producto de estas, t[n]=y[n].y2[n].


```python
n = np.arange(0, 101)

Y1 = np.sin(np.pi * 0.012 * n)

Y2= np.cos(2 * np.pi * 0.03 * n)

s = Y1 + Y2

t = Y1 * Y2
```

m)	Graficar en la misma figura las señales y[n] y y2[n], utilice leyenda y colores diferentes para cada señal. Asigne título a cada eje.

```python
# Crear la figura y los ejes
plt.figure()

# Graficar la primera ecuación con color rojo
plt.plot(n, Y1, color='red', label='y[n]=sin⁡(π*0.12n)')

# Graficar la segunda ecuación con color azul
plt.plot(n, Y2, color='blue', label='y2[n]=cos⁡(2π*0.03n)')

# Añadir las leyendas de los label en la figura
plt.legend()

# Añadir etiquetas a los ejes
plt.xlabel('n')
plt.ylabel('y')

# Añadir un título
plt.title('Graficas de funciones 1 y 2')

# Mostrar el gráfico
plt.show()
```

n)	Graficar en la misma figura las señales s[n] y t[n], utilice leyenda y colores diferentes para cada señal. Asigne título a cada eje.

```python
# Crear la figura y los ejes
plt.figure()

# Graficar la tercera ecuación con color rojo
plt.plot(n, s, color='green', label='s[n]=y[n]+y2[n]')

# Graficar la cuarta ecuación con color violeta
plt.plot(n, t, color='violet', label='y2[n]=cos⁡(2π*0.03n)')

# Añadir las leyendas de los label en la figura
plt.legend()

# Añadir etiquetas a los ejes
plt.xlabel('n')
plt.ylabel('y')

# Añadir un título
plt.title('Graficas de funciones 3 y 4')

# Mostrar el gráfico
plt.show()
```

Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.


```python
def estads_notas(notas):
    notas = pd.Series(notas)
    estadisticas = pd.Series({
        "Nota mínima": notas.min(),
        "Nota máxima": notas.max(),
        "Media": notas.mean(),
        "Desviación estándar": notas.std()
    })
    return estadisticas

# Ejemplo de uso
notas = {
    "Alumno 1": 85,
    "Alumno 2": 90,
    "Alumno 3": 78,
    "Alumno 4": 92,
    "Alumno 5": 88
}

resultado = estads_notas(notas)
print(resultado)
```
